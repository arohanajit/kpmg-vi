```python
import numpy as np
import pandas as pd
import os
```

```python
df = pd.read_excel('kpmg_raw.xlsx', sheet_name="NewCustomerList")
df.columns = df.iloc[0]
df = df.iloc[1:]
df.insert(5,column='age',value=np.nan)

df.head()
```

```python
print(df.columns)
print(len(df.columns))
```

```python
import datetime
for i in df.index:
    if pd.isna(df['DOB'].iloc[i-1]):
        continue
    elif type(df['DOB'].iloc[i-1]) == str:
        df['age'].iloc[i-1] = 2020 - int(df['DOB'].iloc[i-1][:4])
    else:
        df['age'].iloc[i-1] = 2020 - df['DOB'].iloc[i-1].year
    if df['age'].iloc[i-1] >= 100:
        print(df.iloc[i-1])
```

```python
df.isnull().sum()
```

```python
for i in df.index:
    if df['gender'].iloc[i-1][0] == 'M':
        df['gender'].iloc[i-1] = 'Male'
    elif df['gender'].iloc[i-1][0] == 'F':
        df['gender'].iloc[i-1] = 'Female'
    elif df['gender'].iloc[i-1][0] == 'U':
        df['gender'].iloc[i-1] = 'Unknown'
```

```python
df.isnull().sum()
```

```python
df['last_name']=df['last_name'].fillna(method='ffill')
df['DOB']=df['DOB'].fillna(method='ffill')
df['age']=df['age'].fillna(method='ffill')
df['job_title']=df['job_title'].fillna(method='bfill')
df['job_industry_category']=df['job_industry_category'].fillna(method='bfill')
df.columns = df.columns.fillna('to_drop')
df.drop('to_drop', axis = 1, inplace = True)
```

```python
df.isnull().sum()
```

```python
print(max(df['age']),min(df['age']))
```

```python
submission = pd.DataFrame({
        "first_name": df["first_name"],
        "last_name": df["last_name"],
        "gender": df["gender"],
        "past_3_years_bike_related_purchases": df["past_3_years_bike_related_purchases"],
        "DOB": df["DOB"],
        "job_title": df["job_title"],
        "job_industry_category": df["job_industry_category"],
        "wealth_segment": df["wealth_segment"],
        "deceased_indicator": df["deceased_indicator"],
        "owns_car": df["owns_car"],
        "tenure": df["tenure"],
        "address": df["address"],
        "postcode": df["postcode"],
        "state": df["state"],
        "country": df["country"],
        "property_valuation": df["property_valuation"],
        "Rank": df["Rank"],
        "Value": df["Value"]
    })
submission.to_excel('updated_nc.xlsx', index=False)
submission.head(10)
```
